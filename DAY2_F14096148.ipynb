Consider the function $f({\bf x}) = 10x_{0}^{2} + x_{1}^{2} + 5x_{0}x_{1} - 14x_{0} - 6x_{1} +10$.

Define
$$
{\bf x}=
\left[\begin{array}{cc}
x_{0}\\
x_{1}
\end{array}\right],
Q=
\left[\begin{array}{cc}
20&5\\
5&2
\end{array}\right]
\hbox{ and }
{\bf q}=
\left[\begin{array}{cc}
-14\\
-6
\end{array}\right].
$$

Then $f({\bf x})=\displaystyle \frac{1}{2}{\bf x}^tQ{\bf x}+{\bf q}^t {\bf x}+20$. Hence the optimal solution is
$$
{\bf x}^{*}=\left[\begin{array}{cc}
10&4\\
4&2
\end{array}\right]^{-1}\left[\begin{array}{cc}
-14\\
-6
\end{array}\right]=\left[\begin{array}{cc}
1\\
1
\end{array}\right]\hbox{ and } f(1,1)=10.
$$
This function has its optimal solution at
${\bf x}^{*} = (x_{0}^{*},x_{1}^{*}) = (1,1)$ and $f(1,1)=10$.
The descent direction can be
$$
{\bf d}^{k} =
%\left[
%\begin{array}{c}
%    d_{1}^{k} \\
%    d_{2}^{k}
%\end{array}
%\right]=
-\nabla f({\bf x}^k) =
\left[
\begin{array}{c}
    -10x_{0}^{k}-4x_{1}^{k}+14  \\
    -2x_{1}^{k}-4x_{0}^{k}+6
\end{array}
\right]
$$

${\bf Newton’s \quad method}$
<hr>
Suppose we want to solve:
$$
\begin{array}{rl}
\hbox{(P:)} & \displaystyle \min_{x  \in \mathbb{R}} \, f(x ) \\
\end{array}
$$
<hr>
At $x  = \bar{x}$, $f(x)$ can be approximated by:
$$f(x ) \thickapprox h(x ):= f(\bar{x}) + f'(\bar{x})(x -\bar{x}) + \displaystyle \frac{1}{2} (x  - \bar{x}) f''(\bar{x}) (x  - \bar{x}),$$
which is the quadratic Taylor expansion of $f(x )$ at $x  = \bar{x}$. 
<hr>
The minimum of $h(x)$ occur on the set of $h'(x )=0$ is given by 
$$
x=\bar{x}-(f''(x))^{-1}f'(x)(x -\bar{x})
$$
since 
$$
h'(x )=f'(\bar{x})+f''(\bar{x})(x -\bar{x}).
$$
<hr>

    
Step 0: Given ${\bf x}^{0}$, set $k:=0$

Step 1: ${\bf d}^{k}:= -H(x^{k})^{-1}\nabla f({\bf x}^{k})$. If ${\bf d}^{k}=0$, then stop.

Step 2: Choose step-size $\alpha^{k}=1$

Step 3: Set ${\bf x}^{k+1} \leftarrow {\bf x}^{k} + \alpha^{k} {\bf d}^{k}$, $k \leftarrow k+1$.Go to Step 1.

<hr>

Let $f(x) = 7x -\ln{(x)}$. Then
<ul>
    <li>$f'(x) = 7 - \displaystyle \frac{1}{x}$ and $f''(x) = \displaystyle \frac{1}{x^{2}}$.</li>
    <li>$x^{*} =\displaystyle \frac{1}{7} = 0.142857143$ is the unique global minimum.</li>
    <li>The Newton direction at $x^k$ is
$$ d = - (f''(x^k))^{-1} f'(x^k) = -(x^k)^{2} \left( 7 - \displaystyle \frac{1}{x^k} \right) = x^k - 7(x^k)^{2}. $$
    </li>
    <li>Newton’s method will generate the sequence of iterates $\left\{ x^{k} \right\}$ satisfying:
        $$x^{k+1} = x^{k} + (x^{k} - 7(x^{k})^{2}) = 2x^{k} - 7(x^{k})^{2}.$$</li>
<hr>

${\bf Conjugate Gradient Method}$

Example:
$$
\begin{array}{llll}
    4x_{1} & + 3x_{2} &        & = 24\\
    3x_{1} & + 4x_{2} & - x_{3}  & = 30\\
           & - x_{2}  & + 4x_{3} & = -24\\
\end{array}
$$

Define $f(x)=\frac{1}{2}{\bf x}^t Q {\bf x}^t +{\bf q}^t x$. 
$$
Q=\left[
\begin{array}{ccc}
    4 & 3 & 0 \\
    3 & 4 & -1 \\
    0 & -1 & 4 \\
\end{array}
\right],\quad
q=\left[
\begin{array}{c}
    24  \\
    30  \\
    -24 \\
\end{array}
\right]
$$

<hr>
    
Step 0: Compute $r^{1}=-(Qx^{1}+q)$. Set $k=1$ and $d^{1}=r^{1}$.

Step 1: If $r^{k}=0$, Stop. Otherwise, compute
$$
t^{k}=\frac{<r^{k},r^{k}>}{<d^{k},d^{k}>}_{Q} \hbox{ and } x^{k+1}=x^{k}+t^{k}d^{k}
$$

Step 2: Compute $r^{k+1}=r^{k}-t^{k}Qd^{k}$ and $s^{k}=\displaystyle \frac{<r^{k+1},r^{k+1}>}{<r^{k},r^{k}>}$.

Step 3: Compute $d^{k+1}=r^{k+1}+s^{k}d^{k}$. Set $k=k+1$ and goto Step 1.
<hr>

```python
def functionValue(x,Q,q):
    y = x.transpose().dot(Q).dot(x) + q.transpose().dot(x)
    return y
```
